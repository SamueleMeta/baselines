```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
renders = pkl.load(open('../renders.pkl', 'rb'))
```

```python
N = len(renders)
fig, ax = plt.subplots((N//4)+1, 4, sharex=True, sharey=True, figsize=(16, N))
for i, r in enumerate(renders):
    ax[i//4, i%4].imshow(r)
plt.show();
```
