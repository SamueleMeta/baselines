# Basic rendering technique

```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML

rc('animation', html='html5')
```

```python
renders = pkl.load(open('../renders.pkl', 'rb'))
```

```python
fig, ax = plt.subplots(1, figsize=(8, 8))
img = ax.imshow(renders[0])

def init():
    return img

def animate(i):
    img.set_data(renders[i])
    return img

anim = animation.FuncAnimation(fig, animate, frames=len(renders), interval=200);
HTML(anim.to_html5_video())
```
