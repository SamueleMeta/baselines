## CARTPOLE

### Single IS (cannot be normalized)

```python
delta = 0.0001
files = glob.glob(path + '*env=RLLAB.CARTPOLE*norm=none*batchsize=1_*cap=1_*delta=%.4f*' % delta)
IS =  [nu.read_data(logfile) for logfile in files]
```

### MULTIPLE IS (capacity = 10)

```python
delta = 0.0001
files = glob.glob(path + '*env=RLLAB.CARTPOLE*norm=all*batchsize=1_*cap=10_*delta=%.4f*' % delta)
MIS10 =  [nu.read_data(logfile) for logfile in files]
```

### MULTIPLE IS (capacity = 50)

```python
delta = 0.001
files = glob.glob(path + '*env=RLLAB.CARTPOLE*norm=all*batchsize=1_*cap=50_*delta=%.4f*' % delta)
MIS50 =  [nu.read_data(logfile) for logfile in files]
```

### SINGLE IS with batchsize = 10

```python
delta = 0.1
files = glob.glob(path + '*env=RLLAB.CARTPOLE*norm=all*batchsize=10_*cap=1_*delta=%.4f*' % delta)
BATCH =  [nu.read_data(logfile) for logfile in files]
```

### COMPARISON

```python
f = nu.compare({'IS' : IS, 
                'MIS (capacity=10)' : MIS10,
                'MIS (capacity=50)' : MIS50, 
                'IS (batchsize=10)' : BATCH}, key='WindowRewMean', conf=conf, bootstrap=bootstrap, resamples=resamples)
```

```python
f = nu.compare({'IS' : IS, 
                'MIS (capacity=10)' : MIS10,
                'MIS (capacity=50)' : MIS50, 
                'IS (batchsize=10)' : BATCH}, key='WindowRewMean', separate=True, opacity=0.3)
```

```python
f = nu.compare({'IS' : IS, 
                'MIS (capacity=10)' : MIS10,
                'MIS (capacity=50)' : MIS50, 
                'IS (batchsize=10)' : BATCH}, key='WindowRewMean', conf=conf, tolerance=True, prop=prop)
```

```python
nu.save_ti(IS, 'WindowRewMean', conf=conf, prop=prop, rows=501, name='cartpole-ppomis1')
nu.save_ti(MIS10, 'WindowRewMean', conf=conf, prop=prop, rows=501, name='cartpole-ppomis10')
nu.save_ti(MIS50, 'WindowRewMean', conf=conf, prop=prop, rows=501, name='cartpole-ppomis50')
nu.save_ti(BATCH, 'WindowRewMean', conf=conf, prop=prop, rows=501, name='cartpole-batch10')
```
